```python
# Numbreic value by credit Score charts 
y_lst = ['CensusTractMinorityRatioPercent','CensusTractMedFamIncomeAmount','TotalMonthlyIncomeAmount',
         'LoanAcquisitionActualUPBAmt','LTVRatioPercent','LoanAmortizationMaxTermMonths',
         'Borrower1AgeAtApplicationYears','NoteRatePercent','NoteAmount','HousingExpenseRatioPercent',
         'TotalDebtExpenseRatioPercent','PMICoveragePercent', 'ScheduledTotalPaymentCount',
         'LocalAreaMedianIncomeAmount']
for y in y_lst:
    pd_plt = pd.DataFrame(df.groupby('Borrower1CreditScoreValue')[y].mean().T)
    pd_plt.plot.bar()
    plt.show()

# Finding:    
# positive relation with credit score:
# Strong - TotalMonthlyIncomeAmount
# Weak - Borrower1AgeAtApplicationYears

# negative relation with credit score:
# Strong - CensusTractMinorityRatioPercent , LTVRatioPercent, LoanAmortizationMaxTermMonths, TotalDebtExpenseRatioPercent
# mid - HousingExpenseRatioPercent
# Weak -  NoteRatePercent, ScheduledTotalPaymentCount, LocalAreaMedianIncomeAmount

# Mix:
# CensusTractMedFamIncomeAmount, LoanAcquisitionActualUPBAmt, NoteAmount, PMICoveragePercent
```

```python
# Category value credit Score % Chart

y = 'Borrower1GenderType'
df_temp1 =  pd.DataFrame(df.groupby([y, 'Borrower1CreditScoreValue']).size().reset_index(name='count'))
df_temp2 = pd.DataFrame(df.groupby([y]).size().reset_index(name='count'))
df_temp3 = pd.merge(df_temp1, df_temp2, on = y, how = "inner")
df_temp3['percentage_by_cs'] = df_temp3['count_x']/df_temp3['count_y']

df_temp3

for n in range(6):
    print('\033[1m Borrower1CreditScoreValue = '+str(n)+'\033[0m')
    print(df_temp3.loc[df_temp3['Borrower1CreditScoreValue']==n][[y,'percentage_by_cs']].sort_values('percentage_by_cs'))
    pd.DataFrame(df_temp3.loc[df_temp3['Borrower1CreditScoreValue']==n][[y,'percentage_by_cs']].sort_values('percentage_by_cs')).set_index([y]).plot.bar()
    plt.show()

# Desc:    
# 1=Male, 2=Female, 3=Information not provided by borrower, 
# 4=Not Applicable (First or primary borrower is an institution, corporation or partnership), 
# 6=Borrower selected both male and female

# Finding: no relationship 
```
